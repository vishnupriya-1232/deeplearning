```python
import numpy as np
import matplotlib.pyplot as plt

# Metrics for three models
r2_1 = 0.7031567536677323
mse_1 = 318712.73167741473
mae_1 = 117.72871128736203
rmse_1 = 564.5464831857646
r2_2 = 0.7335559022469618
mse_2 = 286073.97097100137
mae_2 = 65.68529145660803
rmse_2 = 534.8588327502888
r2_proposed = 0.8938424231127056
mse_proposed = 113978.57871468436
mae_proposed = 35.0437423119081
rmse_proposed = 337.6071366465531

# List of metrics for each model
metrics_1 = [r2_1, mse_1, mae_1, rmse_1]
metrics_2 = [r2_2, mse_2, mae_2, rmse_2]
metrics_proposed = [r2_proposed, mse_proposed, mae_proposed, rmse_proposed]

# Labels for each metric
labels = ['R^2', 'MSE', 'MAE', 'RMSE']
metrics = np.array([metrics_1, metrics_2, metrics_proposed])
x = np.arange(3)  # 3 models
width = 0.4

# Plot for R^2
plt.figure(figsize=(6, 4))
plt.bar(x, [r2_1, r2_2, r2_proposed], width, color=['blue', 'orange', 'green'])
plt.title('R^2')
plt.xticks(x, ['ANN', 'CNN', 'RNN'])
plt.show()

```

```python

# Plot for MSE
plt.figure(figsize=(6, 4))
plt.bar(x, [mse_1, mse_2, mse_proposed], width, color=['blue', 'orange', 'green'])
plt.title('MSE')
plt.xticks(x, ['ANN', 'CNN', 'RNN'])
plt.show()
```

```python
# Plot for MAE
plt.figure(figsize=(6, 4))
plt.bar(x, [mae_1, mae_2, mae_proposed], width, color=['blue', 'orange', 'green'])
plt.title('MAE')
plt.xticks(x, ['ANN', 'CNN', 'RNN'])
plt.show()
```

```python
# Plot for RMSE
plt.figure(figsize=(6, 4))
plt.bar(x, [rmse_1, rmse_2, rmse_proposed], width, color=['blue', 'orange', 'green'])
plt.title('RMSE')
plt.xticks(x, ['ANN', 'CNN', 'RNN'])
plt.show()
```
